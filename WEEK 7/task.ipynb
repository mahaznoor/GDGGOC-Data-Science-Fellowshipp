# Unsupervised Learning and Market Basket Analysis

This notebook demonstrates different unsupervised learning techniques including:

- K-Means Clustering (implemented from scratch)
- Hierarchical Clustering
- Dimensionality Reduction using PCA
- Data Visualization using t-SNE
- Association Rule Mining using Apriori

The final part of this project focuses on Market Basket Analysis using association rule mining to discover relationships between items purchased together in a retail store.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

plt.scatter(X[:,0], X[:,1])
plt.title("Original Dataset")
plt.show()
```

## K-Means Clustering (From Scratch)

K-Means is an unsupervised learning algorithm used to group similar data points into clusters.

Steps:
1. Choose number of clusters K
2. Randomly initialize centroids
3. Assign points to nearest centroid
4. Update centroid positions
5. Repeat until convergence

```python
def kmeans(X, k, iterations=100):

    centroids = X[np.random.choice(len(X), k, replace=False)]

    for _ in range(iterations):

        clusters = [[] for _ in range(k)]

        for point in X:
            distances = [np.linalg.norm(point-centroid) for centroid in centroids]
            cluster_index = np.argmin(distances)
            clusters[cluster_index].append(point)

        new_centroids = []

        for cluster in clusters:
            new_centroids.append(np.mean(cluster, axis=0))

        centroids = np.array(new_centroids)

    return centroids, clusters

centroids, clusters = kmeans(X,3)
```

```python
for cluster in clusters:
    cluster = np.array(cluster)
    plt.scatter(cluster[:,0], cluster[:,1])

plt.scatter(centroids[:,0], centroids[:,1], c='red', s=200)
plt.title("K-Means Clustering")
plt.show()
```

## Hierarchical Clustering

Hierarchical clustering builds nested clusters by merging or splitting them successively.

Types:
- Agglomerative (bottom-up)
- Divisive (top-down)

In this notebook we use Agglomerative Clustering.

```python
model = AgglomerativeClustering(n_clusters=3)

labels = model.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Hierarchical Clustering")
plt.show()
```

## Principal Component Analysis (PCA)

PCA is a dimensionality reduction technique that transforms high-dimensional data into fewer dimensions while retaining maximum variance.

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:,0], X_pca[:,1])
plt.title("PCA Result")
plt.show()
```

## t-SNE Visualization

t-SNE is a powerful technique used to visualize high dimensional data in 2D or 3D.

```python
tsne = TSNE(n_components=2)

X_tsne = tsne.fit_transform(X)

plt.scatter(X_tsne[:,0], X_tsne[:,1])
plt.title("t-SNE Visualization")
plt.show()
```

## Market Basket Analysis using Apriori

Market Basket Analysis is used to find associations between products purchased together.

Example:
Customers who buy Bread and Butter also buy Milk.

```python
data = {
'Milk':[1,0,1,1,0],
'Bread':[1,1,1,0,0],
'Butter':[0,1,1,1,0],
'Eggs':[1,1,0,1,1]
}

df = pd.DataFrame(data)

df
```

```python
frequent_items = apriori(df, min_support=0.4, use_colnames=True)

frequent_items
```

```python
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.6)

rules
```

```python
sns.scatterplot(x="support", y="confidence", data=rules)
plt.title("Association Rules")
plt.show()
```
